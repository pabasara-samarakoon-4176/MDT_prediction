```python
def positional_encoding(seq_len, d_model):
    pos = torch.arange(seq_len, dtype=torch.float32).unsqueeze(1)
    i = torch.arange(d_model, dtype=torch.float32).unsqueeze(0)
    angle_rates = 1 / torch.pow(10000, (2 * (i // 2)) / d_model)
    angle_rads = pos * angle_rates

    pos_encoding = torch.zeros(seq_len, d_model)
    pos_encoding[:, 0::2] = torch.sin(angle_rads[:, 0::2])
    pos_encoding[:, 1::2] = torch.cos(angle_rads[:, 1::2])
    return pos_encoding
```

```python
import geohash2

def geohash_to_position_tensor(geohash_list):
    """Convert a list of geohashes to positional encoding via lat/lon"""
    positions = [(float(lat), float(lon)) for lat, lon in [geohash2.decode(g) for g in geohash_list]]
    pos_tensor = torch.tensor(positions, dtype=torch.float32)  # [N, 2]
    return pos_tensor
```

```python
# complete model
import torch
from torch import nn
import torch.nn.functional as F
import math

class RSRPTransformerModel(nn.Module):
    def __init__(self, input_dim=11, d_model=64, d_ff=2, output_dim=1):
        super(RSRPTransformerModel, self).__init__()
        self.d_model = d_model
        self.input_proj = nn.Linear(input_dim, d_model)

        # Positional Encoding from Geohash (Lat, Lon → d_model)
        self.position_proj = nn.Linear(2, input_dim)

        # Self-attention weights
        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)

        # LayerNorm
        self.layer_norm1 = nn.LayerNorm(d_model)
        self.layer_norm2 = nn.LayerNorm(d_model)

        # Feedforward network
        self.ffn = nn.Sequential(
            nn.Linear(d_model, d_ff),
            nn.ReLU(),
            nn.Linear(d_ff, d_model)
        )

        # Output layer for RSRP prediction
        self.output_layer = nn.Linear(d_model, output_dim)

    def forward(self, x, geohash_pos_tensor):
        # Positional encoding projection
        pos_embed = self.position_proj(geohash_pos_tensor)
        x = x + pos_embed  # Inject position (geohash-derived) info

        # Add sequence dimension
        x = x.unsqueeze(1)  # Shape: [N, 1, input_dim]

        # Project input to d_model
        x_proj = self.input_proj(x)  # Shape: [N, 1, d_model]

        # Compute attention
        Q = self.W_q(x_proj)
        K = self.W_k(x_proj)
        V = self.W_v(x_proj)

        attention_scores = Q @ K.transpose(-2, -1) / math.sqrt(self.d_model)
        attention_weights = F.softmax(attention_scores, dim=-1)
        attention_output = attention_weights @ V

        # Residual + Norm
        res_1 = self.layer_norm1(x_proj + attention_output)

        # Feedforward
        ffn_output = self.ffn(res_1)
        res_2 = self.layer_norm2(res_1 + ffn_output)

        # Squeeze sequence dimension
        res_2 = res_2.squeeze(1)  # Shape: [N, d_model]

        # Predict RSRP
        return self.output_layer(res_2)
```

```python
model = RSRPTransformerModel(input_dim=11, d_model=64)
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    df, test_size=0.2, random_state=42
)
```

```python
# Training
x = torch.tensor(train_df[features + ['Geohash_index']].values, dtype=torch.float32)
geohash_pos_tensor = geohash_to_position_tensor(train_df['Geohash'].tolist())
target_tensor = torch.tensor(train_df['RSRP_norm'].values, dtype=torch.float32).view(-1, 1)
```

```python
model = RSRPTransformerModel(input_dim=11, d_model=64)
```

```python
import torch.nn as nn
import torch.optim as optim

loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
num_epochs = 100

for epoch in range(num_epochs):
  model.train()
  # Forward pass
  preds = model(x, geohash_pos_tensor)
  loss = loss_fn(preds, target_tensor)
  # Backward pass
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
  if (epoch + 1) % 10 == 0 or epoch == 0:
      print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
# Features and target
features = ['Latitude', 'Longitude', 'Azimuth', 'Tilt', 'Height',
            'Distance_to_cell', 'Avg_Angle_with_cell', 'NDVI',
            'RoadDensity', 'Elevation']

# Use the scaler already fitted on `train_df`
X_test = test_df[features]
geohash_positions_test = geohash_to_position_tensor(test_df['Geohash'].tolist())

# Convert to tensors
input_tensor_test = torch.tensor(
    np.hstack([X_test, test_df[['Geohash_index']].values]), dtype=torch.float32
)
geohash_pos_test = geohash_positions_test

# Target tensor (normalized)
target_tensor_test = torch.tensor(
    test_df['RSRP_norm'].values, dtype=torch.float32).view(-1, 1)
```

```python
model.eval()
with torch.no_grad():
    preds_test = model(input_tensor_test, geohash_pos_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Denormalize if needed
rsrp_pred_denorm = rsrp_scaler.inverse_transform(preds_test.numpy())
target_denorm = rsrp_scaler.inverse_transform(target_tensor_test.numpy())

# Metrics
mse = mean_squared_error(target_denorm, rsrp_pred_denorm)
mae = mean_absolute_error(target_denorm, rsrp_pred_denorm)
r2 = r2_score(target_denorm, rsrp_pred_denorm)

print(f"Test MSE: {mse:.4f}")
print(f"Test MAE: {mae:.4f}")
print(f"Test R²:  {r2:.4f}")
```
