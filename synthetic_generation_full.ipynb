```python
import numpy as np
from geopy.distance import geodesic
from geopy.point import Point
import random
```

```python
antenna_height_range = (df_without_terrain["Antenna Height"].min(), df_without_terrain["Antenna Height"].max())
azimuth_range = (df_without_terrain["Azimuth"].min(), df_without_terrain["Azimuth"].max())
tilt_range = (df_without_terrain["Tilt"].min(), df_without_terrain["Tilt"].max())
```

```python
earfcn_values = [1650, 525, 3750]
earfcn_to_max_radius = {1650: 1000, 525: 3000, 3750: 500}
```

```python
lat_min, lat_max = 5.9, 9.9
lon_min, lon_max = 79.7, 81.9
```

```python
def generate_mdt_point(site_lat, site_lon, radius_m):
    angle = np.random.uniform(0, 2 * np.pi)
    distance_km = np.random.uniform(0, radius_m / 1000)
    origin = Point(site_lat, site_lon)
    destination = geodesic(kilometers=distance_km).destination(origin, np.degrees(angle))
    return destination.latitude, destination.longitude
```

```python
from tqdm import tqdm

def generate_synthetic_mdt(num_cells):
    records = []
    for cell_id in tqdm(range(1000, 1000 + num_cells)):
        site_lat = np.random.uniform(lat_min, lat_max)
        site_lon = np.random.uniform(lon_min, lon_max)

        earfcn = random.choice(earfcn_values)
        antenna_height = np.random.uniform(*antenna_height_range)
        azimuth = np.random.uniform(*azimuth_range)
        tilt = np.random.uniform(*tilt_range)

        is_urban = np.random.rand() < 0.6
        num_points = np.random.randint(100, 250) if is_urban else np.random.randint(10, 40)

        for _ in range(num_points):
            mdt_lat, mdt_lon = generate_mdt_point(site_lat, site_lon, earfcn_to_max_radius[earfcn])

            rsrp = np.random.normal(df_without_terrain['RSRP'].mean(), df_without_terrain['RSRP'].std())
            rsrq = np.random.normal(df_without_terrain['RSRQ'].mean(), df_without_terrain['RSRQ'].std())

            record = {
                "Cell_ID": f"CELL_{cell_id}",
                "RSRP": rsrp,
                "RSRQ": rsrq,
                "EARFCN_DL": earfcn,
                "Latitude_of_MDT": mdt_lat,
                "Longitude_of_MDT": mdt_lon,
                "Antenna Height": antenna_height,
                "Azimuth": azimuth,
                "Tilt": tilt,
                "Latitude_of_Site": site_lat,
                "Longitude_of_Site": site_lon
            }
            records.append(record)
    return pd.DataFrame(records)
```

```python
synthetic_df = generate_synthetic_mdt(num_cells=2000)
```

```python
import rasterio
from rasterio.sample import sample_gen

def extract_raster_values(gdf, raster_path, value_column):
    with rasterio.open(raster_path) as src:
        gdf_proj = gdf.to_crs(src.crs)
        coords = [(geom.x, geom.y) for geom in gdf_proj.geometry]
        values = [val[0] if val[0] is not None else None for val in src.sample(coords)]
        gdf[value_column] = values
    return gdf
```

```python
synthetic_gdf = extract_raster_values(synthetic_gdf, elevation_raster_file, 'elevation')
synthetic_gdf = extract_raster_values(synthetic_gdf, slope_raster_file, 'slope')
```

```python
# NDVI
```

```python
ndvi_west = "/content/drive/MyDrive/Final_year_project/vegetation/ndvi_west.tiff"
ndvi_east = "/content/drive/MyDrive/Final_year_project/vegetation/ndvi_east.tiff"

ndvi_files = [ndvi_west, ndvi_east]
```

```python
from rasterio.transform import rowcol

synthetic_df['NDVI'] = np.nan
```

```python
rasters = []
for file in ndvi_files:
    src = rasterio.open(file)
    data = src.read(1).astype(np.float32) / 65535.0
    rasters.append((src, data))
```

```python
def get_ndvi(lat, lon):
    for src, ndvi_data in rasters:
        try:
            row, col = rowcol(src.transform, lon, lat)
            if (0 <= row < ndvi_data.shape[0]) and (0 <= col < ndvi_data.shape[1]):
                return float(ndvi_data[row, col])
        except:
            continue
    return np.nan
```

```python
synthetic_df["NDVI"] = synthetic_df.progress_apply(lambda row: get_ndvi(row["Latitude_of_MDT"], row["Longitude_of_MDT"]), axis=1)
```
