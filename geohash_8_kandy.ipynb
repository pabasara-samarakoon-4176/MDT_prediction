```python
import pandas as pd
import numpy as np
import random
import geohash2
import geopy.distance
```

```python
# np.random.seed(42)
# random.seed(42)
```

```python
def generate_geohashes(num_geohashes):
    # Choose a central point — e.g., around Kandy, Sri Lanka
    center_lat = 7.2906
    center_lon = 80.6337

    # Estimate how many grid points per axis (square grid)
    side = int(np.ceil(np.sqrt(num_geohashes)))

    # Use a small range for clustering — e.g., ±0.005 degrees (~500m)
    lat_step = 0.00018
    lon_step = 0.00018

    # Create grid offsets
    lat_grid = center_lat + lat_step * (np.arange(side) - side // 2)
    lon_grid = center_lon + lon_step * (np.arange(side) - side // 2)

    # Create meshgrid of lat/lon combinations
    lats, lons = np.meshgrid(lat_grid, lon_grid)
    lats = lats.flatten()[:num_geohashes]
    lons = lons.flatten()[:num_geohashes]

    geohashes = [geohash2.encode(lat, lon, precision=8) for lat, lon in zip(lats, lons)]
    return geohashes, lats, lons
```

```python
num_rows = 512
geohashes, lats, lons = generate_geohashes(num_rows)
```

```python
def generate_features(n):
    Azimuth = np.random.uniform(0, 360, n)
    Tilt = np.random.uniform(-10, 10, n)
    Height = np.random.uniform(10, 60, n)
    Distance_to_cell = np.random.uniform(10, 500, n)
    Avg_Angle_with_cell = np.random.uniform(0, 90, n)
    NDVI = np.random.uniform(0, 0.9, n)
    RoadDensity = np.random.uniform(0, 1, n)
    Elevation = np.random.uniform(80, 300, n)
    base_rsrp = (
        10 * RoadDensity - 15 * NDVI - 0.05 * Distance_to_cell + np.random.normal(0, 3, n) - 95
    )
    return {
        "Geohash": geohashes,
        "Latitude": lats,
        "Longitude": lons,
        "Azimuth": Azimuth,
        "Tilt": Tilt,
        "Height": Height,
        "Distance_to_cell": Distance_to_cell,
        "Avg_Angle_with_cell": Avg_Angle_with_cell,
        "NDVI": NDVI,
        "RoadDensity": RoadDensity,
        "Elevation": Elevation,
        "RSRP": base_rsrp
    }
```

```python
data = pd.DataFrame(generate_features(num_rows))
data.head()
```

```python
from shapely.geometry import Polygon
import geopandas as gpd
```

```python
def geohash_to_polygon(gh):
    lat, lon, lat_err, lon_err = geohash2.decode_exactly(gh)
    lat_min, lat_max = lat - lat_err, lat + lat_err
    lon_min, lon_max = lon - lon_err, lon + lon_err
    return Polygon([
        (lon_min, lat_min),
        (lon_min, lat_max),
        (lon_max, lat_max),
        (lon_max, lat_min),
        (lon_min, lat_min)
    ])
```

```python
data['geometry'] = data['Geohash'].apply(geohash_to_polygon)
gdf = gpd.GeoDataFrame(data, geometry='geometry')
```

```python
gdf.head()
```

```python
import folium

center_lat = gdf.geometry.centroid.y.mean()
center_lon = gdf.geometry.centroid.x.mean()
m1 = folium.Map(location=[center_lat, center_lon], zoom_start=13)

for _, row in gdf.iterrows():
    color = 'red' if row['RSRP'] < -100 else 'green' if row['RSRP'] > -90 else 'orange'

    folium.GeoJson(
        row['geometry'].__geo_interface__,
        style_function=lambda feature, color=color: {
            'fillColor': color,
            'color': 'black',
            'weight': 1,
            'fillOpacity': 0.6
        },
        tooltip=folium.Tooltip(f"Geohash: {row['Geohash']}<br>RSRP: {row['RSRP']:.2f}")
    ).add_to(m1)
```

```python
import branca.colormap as cm

colormap = cm.linear.RdYlGn_11.scale(gdf['RSRP'].min(), gdf['RSRP'].max())
colormap.caption = 'RSRP (dBm)'
colormap.add_to(m1)
```
