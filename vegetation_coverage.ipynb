```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import geohash2
from pyrosm import OSM
```

```python
df = pd.read_csv('/content/drive/MyDrive/Final_year_project/datasets/Sri-Lanka.csv')
df.head()
```

```python
df = df.drop(columns = ['Geohash'])
```

```python
df['geometry'] = df.apply(lambda row: Point(row['Longitude_of_MDT'], row['Latitude_of_MDT']), axis=1)
df['geohash'] = df.apply(
    lambda row: geohash2.encode(row['Latitude_of_MDT'], row['Longitude_of_MDT'], precision=8), axis=1
)
```

```python
df.head()
```

```python
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")
```
